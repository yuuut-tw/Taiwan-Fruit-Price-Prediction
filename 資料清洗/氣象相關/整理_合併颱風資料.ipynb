### 合併入分析用資料中
- 影響菜價時間暫定為前後3天
- 強度轉為數值 => 0~3(弱～強)

```python
# 讀取2011年以後的颱風資料
df_typhoon = pd.read_csv("cleaned data/typhoon_tmp.csv")
df_typhoon.head()
```

```python
# 給予編號 => 稍後與影響時間df合併
df_typhoon["編號"] = df_typhoon.index
df_typhoon.head()
```

```python
# 將菜價影響時間暫定為正負3天，共5天 => 前3天為預期心理，後3天為影響時間
# 根據研究指出，一般會影響價格達數週 => https://www.coa.gov.tw/ws.php?id=22086&print=Y
time = []
num = []

n = 0
ty_time = df_typhoon["海上警報開始時間"].tolist()
for i in ty_time:
    x = pd.date_range(end=i, periods=3).tolist()
    y = pd.date_range(start=i, periods=3).tolist()
    output = sorted(list(set(x+y)), reverse=True) # 除去同天值，在排序
    time += output
    num += [str(n)]*len(output)
    n += 1 

# create df
df_test = pd.DataFrame({"日期": time,
                        "編號": num
                       })
df_test.head(10)
```

```python
# 將編號由字串轉為int => 應該有更好寫法
df_test["編號"] = df_test["編號"].astype("int")
```

```python
# 合併＆刪除一些不必要欄位
df1 = pd.merge(df_typhoon, df_test, on="編號", how="left").drop(columns=["英文名稱", "海上警報開始時間", "編號"])
df1
```

```python
# 讀取分析資料
df2 = pd.read_csv("analysis data/台中_香蕉合併資料(價格&天氣).csv")
df2
```

```python
df2["日期"] = pd.to_datetime(df2["日期"])
```

```python
# 合併
df3 = pd.merge(df2, df1, on="日期", how="left")
```

```python
df3.tail(10)
```

```python
# 新增欄位：有颱風為1，無颱風為0 => 取代名稱（暫）
df3["typhoon"] = df3["警報發布報數"].apply(lambda x: 1 if x>0 else 0) 
```

```python
# 將沒有颱風時的近臺強度,警報發布報數改為0，然後有颱風的近臺強度用np.where把強度以數值替代
df3.loc[~(df3["警報發布報數"] > 0), "警報發布報數"] = 0
```

```python
df3["近臺強度"].unique()
```

```python
condition = [
    (df3["近臺強度"] == np.nan),
    (df3["近臺強度"] == "w"),
    (df3["近臺強度"] == "m"),
    (df3["近臺強度"] == "s"),
]
result = [0, 1,2,3]

df3["近臺強度"] = np.select(condition, result)
```
