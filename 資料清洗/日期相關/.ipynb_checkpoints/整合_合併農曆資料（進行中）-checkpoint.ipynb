```python
import pandas as pd
import re
```

```python
df_cal = pd.read_csv("../data/Calender.csv")
```

```python
df_cal.head(15)
```

```python
df_cal["SolarCalendar"] = pd.to_datetime(df_cal["SolarCalendar"], format="%Y-%m-%d")
```

```python
num_dict = {'一':1, '二':2, '三':3, '四':4, '五':5,
            '六':6, '七':7, '八':8, '九':9, '十':10}

def lunar_day(text):
    if text[0] == "初":
        return num_dict[text[1]]
    elif text[0] == "十":
        return 10 + num_dict[text[1]]
    elif text[0] == "廿":
        return 20 + num_dict[text[1]]
    elif text in ["三十", "二十"]:
        return num_dict[text[0]] * 10
```

### 轉化農曆日為數值

```python
df_cal["LunarDay"] = df_cal["LunarCalendar"].apply(lambda x: re.findall("月(.*)日", x)[0])
df_cal["LunarDay"] = df_cal["LunarDay"].apply(lunar_day)
```
