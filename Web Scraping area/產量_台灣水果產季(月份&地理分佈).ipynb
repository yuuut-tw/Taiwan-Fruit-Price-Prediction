```python
import requests
import pandas as pd
import json
```

```python
url = "https://data.coa.gov.tw/Service/OpenData/DataFileService.aspx?UnitId=061&$top=6000&$skip=0" # "https://data.coa.gov.tw/Service/OpenData/DataFileService.aspx?UnitId=113" #"https://data.coa.gov.tw/Service/OpenData/FromM/TownCropData.aspx"
headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'
            }

res = requests.get(url, headers=headers)
data = json.loads(res.text)
```

```python
data
```

### 將json檔案轉換成dataframe

```python
df = pd.json_normalize(data)
df
```

### data cleaning
- 建立水果類df => df_fruit
- 拿掉不必要欄位

```python
df_fruit = df.loc[df["type"] == '水果']
df_fruit
```

### 轉換data type
[SOF - Change column type in pandas](https://stackoverflow.com/questions/15891038/change-column-type-in-pandas)

```python
df_fruit["month"] = df_fruit["month"].astype("int")
```

```python
df_fruit
```

```python
df_fruit.to_csv("台灣水果產季(月份&分布).csv", index=False)
```
